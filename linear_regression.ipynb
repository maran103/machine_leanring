```python
import numpy as np
import pandas as pd
from sklearn import linear_model as lm
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/sound/Documents/homeprices.csv")
print(df.head())
```

```python
reg = lm.LinearRegression()
reg.fit(df[['area']],df[['price']])
```

```python
a = reg.predict([[4000]])
print(a)
b = reg.score(df[['area']],df[['price']])
print(b)
print(df.shape)
```

```python
df['another'] = [3000,4500,6000,4000,5500]
print(df.head())
plt.scatter(df.area,df.price,color='red',marker="+")
#plt.plot(df.area,reg.predict(df[['area']],),color='green')
plt.plot(df.area,reg.predict(df[['area']]),color='yellow')
plt.plot(df.area,reg.predict(df[['another']]))
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```
