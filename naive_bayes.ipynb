```python
import pandas as pd
df = pd.read_csv("C:/Users/sound/Downloads/titanic.csv")
print(df.head())
df.drop(['PassengerId','Name','SibSp','Cabin','Embarked','Ticket','Parch'],axis='columns',inplace=True)
print(df.head())
target = df['Survived']
df1 = df.drop(['Survived'],axis='columns')
print(df1.head())
dummies = pd.get_dummies(df1.Sex)
print(dummies)
df1 = pd.concat([df1,dummies],axis='columns')
print(df1.head())
df1.drop('Sex',axis='columns',inplace=True)
a=df1['Age'].mean()
print(a)
df1['Age'] = df1['Age'].fillna(a)
print(df1[:10])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
a = model.score(X_test,y_test)
print(a)
print(model.predict(X_test[:10]))
print(model.predict_proba(X_test[:10]))
print(X_train[:10])
print(X_test[:10])
print(y_train[:10])
print(y_test[:10])
print(model.predict(X_test[:10]))
```
