```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/Melbourne_housing_FULL.csv')
df.head()
```

```python
df['Price'] = df['Price'].fillna(df['Price'].mean())
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
```

```python
df['Landsize']=df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea']=df['BuildingArea'].fillna(df['BuildingArea'].mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df = pd.get_dummies(df,drop_first=True)
df.head()
```

```python
x =  df.drop('Price',axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

```python
from sklearn import linear_model

lesso_reg = linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)

lesso_reg.fit(x_train,y_train)
```

```python
lesso_reg.score(x_test,y_test)
```

```python
lesso_reg.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
ridge_reg.score(x_train,y_train)
```
