```python
import pandas as pd
import numpy as np
```

price = m1*area + m2*bedrooms + m3*age + b<br>
m1,m2,m3 = coefficient b = intercept ,price is a dependent variable area , bedrooms, age are independent

```python
df = pd.read_csv("C:/Users/sound/Documents/homeprices2.csv")
print(df)
```

```python
a = df['bedrooms'].median()
print(a)
b = int(a)
```

```python
df['bedrooms'] = df.bedrooms.fillna(b)
print(df['bedrooms'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(df[['area','bedrooms','age']],df['price'])
```

```python
predicted_value = model.predict([[3000,3,40]])
print(predicted_value)
```

```python
print(model.coef_)
print(model.intercept_)
print(model.score(df[['area','bedrooms','age']],df['price']))
```
